```python
import numpy as np
import matplotlib.pyplot  as plt
import seaborn as sns
import pandas as pd

class NeuralNetwork(object):
    def __init__(self,NN_ARCHITECTURE):

        self.NN_ARCHITECTURE = NN_ARCHITECTURE
        # Random seed initiation
        seed = 10;
        np.random.seed(seed);

        # no. of layers
        no_of_layers = len(NN_ARCHITECTURE);

        self.params_values ={}
        self.weight_history = []
        # iteration over network layers
        for idx,layer in enumerate(NN_ARCHITECTURE):
            layer_idx = idx+1
            # extracting no, of units in layer
            layer_ip_size = layer["input_dim"]
            layer_op_size = layer["output_dim"]

            # initiating weight matrix and bias for subsequent layers
            self.params_values['W'+str(layer_idx)] = np.random.randn(layer_op_size, layer_ip_size)*0.1
            self.params_values['b' + str(layer_idx)] = 0#np.random.randn(layer_op_size, 1) * 0.1
            if idx==0:
                n=np.array(self.params_values['W'+str(layer_idx)])
            else:
                n = np.append(n,self.params_values['W'+str(layer_idx)])
        print("Initial Weights",self.params_values['W1'])
        self.weight_history = np.array(n)
        # return param_values

    def sigmoid(self,Z):
        # function returning sigmoid output of Z
        return 1 / (1 + np.exp(-Z))
    
    def htan(self,Z):
          return (np.exp(Z) - np.exp(-Z))/(np.exp(Z) + np.exp(-Z))


    def feedforward(self,x,params_values):
        # calculation of the input value for the activation function
        memory ={}
        # X vector is activation for layer 0
        A_curr = x
        # iteration over network layers
        for idx, layer in enumerate(self.NN_ARCHITECTURE):
            # we number network layers from 1
            layer_idx = idx + 1
            # transfer the activation from the previous iteration
            A_prev = A_curr

            # extraction of the activation function for the current layer
            activ = layer["activation"]
            # extraction of W for the current layer
            W = params_values["W" + str(layer_idx)]
            # extraction of b for the current layer
            b = params_values["b" + str(layer_idx)]
            # calculation of activation for the current layer
            z = np.dot(W, A_prev) + b
           # print(str(z))
        
            activ = input()
        
            # selection of activation function
            if activ == "htan":
                A_curr = self.htan(z)
            elif activ == "sigmoid":
                A_curr = self.sigmoid(z)
            else:
                raise Exception('Non-supported activation function')

            print("A_curr",A_curr)
            # saving calculated values in the memory
            memory["A" + str(idx)] = A_prev
            memory["Z" + str(layer_idx)] = z

        # return of prediction vector and a dictionary containing intermediate values
        return A_curr, memory

    def get_cost_value(self,Y_hat, Y):
        # number of examples
        #print("Cost",Y_hat)
        m = Y_hat.shape[1]
        # calculation of the cost according to the formula
        cost = -1 / m * (np.dot(Y, np.log(Y_hat).T) + np.dot(1 - Y, np.log(1 - Y_hat).T))
#         category=np.argmax(Y_hat,axis=1)
#         print(category,Y)
#         loss=((category-np.transpose(Y))**2).mean()
        
        #print("Cost",np.log(Y_hat).T)
        return np.squeeze(cost) #np.squeeze(loss) 

    # an auxiliary function that converts probability into class
    def convert_prob_into_class(self,probs):
        probs_ = np.copy(probs)
        probs_[probs_ > 0.5] = 1
        probs_[probs_ <= 0.5] = 0
        return probs_

    def get_accuracy_value(self,Y_hat, Y):
        #print("Tha accuracy value", Y_hat, Y)
        Y_hat_ = self.convert_prob_into_class(Y_hat)
        #print("Tha accuracy value",Y_hat_,Y)
#         category=np.argmax(Y_hat,axis=1)
#         accuracy=(category==Y).mean()
        return (Y_hat_ == Y).all(axis=0).mean()
    def backward(self,Y_hat, Y, memory):
        grads_values = {}

        # number of examples
        m = Y.shape[1]
        # a hack ensuring the same shape of the prediction vector and labels vector
        Y = Y.reshape(Y_hat.shape)

        # initiation of gradient descent algorithm
        dA_prev = - (np.divide(Y, Y_hat) - np.divide(1 - Y, 1 - Y_hat));
        #dA_prev = 2*(Y_hat -Y)/Y_hat #mean bias error
        for layer_idx_prev, layer in reversed(list(enumerate(self.NN_ARCHITECTURE))):
            # we number network layers from 1
            layer_idx_curr = layer_idx_prev + 1
            # extraction of the activation function for the current layer
            activ = layer["activation"]

            dA_curr = dA_prev

            A_prev = memory["A" + str(layer_idx_prev)]
            Z_curr = memory["Z" + str(layer_idx_curr)]

            W_curr = self.params_values["W" + str(layer_idx_curr)]
            b_curr = self.params_values["b" + str(layer_idx_curr)]

            m = A_prev.shape[1]
            
            activ = input()
            
            # selection of activation function
            if activ == "htan":
                dZ_curr = self.relu_backward(Z_curr)
            elif activ == "sigmoid":
                dZ_curr = self.sigmoid_backward(dA_curr, Z_curr)
            else:
                raise Exception('Non-supported activation function')

            # derivative of the matrix W
            dW_curr = np.dot(dZ_curr, A_prev.T) / m
            # derivative of the vector b
            db_curr = np.sum(dZ_curr, axis=1, keepdims=True) / m
            # derivative of the matrix A_prev
            dA_prev = np.dot(W_curr.T, dZ_curr)
            grads_values["dW" + str(layer_idx_curr)] = dW_curr
            grads_values["db" + str(layer_idx_curr)] = db_curr
        return grads_values

    def update(self,params_values, grads_values, learning_rate):

        # iteration over network layers
        for layer_idx, layer in enumerate(self.NN_ARCHITECTURE, 1):
            params_values["W" + str(layer_idx)] -= learning_rate * grads_values["dW" + str(layer_idx)]
            params_values["b" + str(layer_idx)] -= learning_rate * grads_values["db" + str(layer_idx)]
            #self.weight_history[layer_idx-1].append(params_values["W" + str(layer_idx)])
            if layer_idx==1:
                n=np.array(self.params_values['W'+str(layer_idx)])
            else:
                n = np.append(n,self.params_values['W'+str(layer_idx)])
        self.weight_history = np.vstack((self.weight_history,n))
        return params_values;

    def train(self,X, Y, epochs, learning_rate,batch,cl, verbose=False, callback=None):
        # initiation of neural net parameters
        params_values = self.params_values
        # initiation of lists storing the history
        # of metrics calculated during the learning process
        cost_history = []
        accuracy_history = []
        weight_history =[self.params_values]
        # performing calculations for subsequent iterations
        targets = np.zeros((len(Y),cl), np.float32)
        targets[range(targets.shape[0]),Y] = 1
        targets = np.transpose(targets)
        for i in range(epochs):
            # step forward
            Y_hat, cashe = self.feedforward(X,self.params_values)
            # calculating metrics and saving them in history
            cost = self.get_cost_value(Y_hat, Y)
            cost_history.append(cost)
            accuracy = self.get_accuracy_value(Y_hat, Y)
            accuracy_history.append(accuracy)

            # step backward - calculating gradient
            grads_values = self.backward(Y_hat, Y, cashe)
            # updating model state
            self.params_values = self.update(self.params_values, grads_values,learning_rate)
            weight_history.append(self.params_values)
            print("Weight values:",str(self.params_values))
            if (i % 50 == 0):
                #print("Weight values:",str(self.params_values))
                #print(i)
                #if (verbose):
                print("Iteration: {:05} - cost: {:.5f} - accuracy: {:.5f}".format(i, cost, accuracy))
                if (callback is not None):
                    callback(i, params_values)
        print(self.weight_history)
        plt.figure()
        plt.subplot(211)
        plt.plot(list(range(epochs+1)), self.weight_history)# plot the cost function.
        plt.ylabel('Edge Weights')
        plt.xlabel('Iterations')
        plt.subplot(212)
        plt.plot(list(range(epochs)), accuracy_history, '-r')# plot the cost function.
        plt.ylabel('Accuracy')
        plt.xlabel('Iterations')

        
        return params_values
    
NN = NeuralNetwork(NN_ARCHITECTURE)
```

```python
NN_ARCHITECTURE = [
    {"input_dim": 2, "output_dim": 10, "activation": "htan"},
    {"input_dim": 10, "output_dim": 10, "activation": "htan"},
    {"input_dim": 10, "output_dim": 1, "activation": "sigmoid"}
]
```

```python
# Dataset visualization
train_file_name = 'data.simple.test.1000.csv'
batch_size = int(input('batch size:'))
epoch = int(input('epoch:'))
learning_rate = int(input('learning rate:'))
train_ip = pd.read_csv(train_file_name)
train_ip.head()
sns.FacetGrid(train_ip, hue="smoking", height=5) \
   .map(plt.scatter, "Imaginary part", "Real part") \
   .add_legend()
```

```python
#initate class object
NN = NeuralNetwork(NN_ARCHITECTURE)

train_ip = np.loadtxt(fname=train_file_name,delimiter=',',skiprows=1,max_rows=batch_size)

# seggregate input and output
y = np.array(train_ip[:,2],dtype=int)
y = np.where(y==1,0,y)
y = np.where(y==2,1,y)
X = np.array(train_ip[:,:2],dtype=float)
#print(str(X),str(np.transpose(y)))
# Training
params_values = NN.train(np.transpose(X), np.transpose(y.reshape((y.shape[0], 1))), epoch, learning_rate,batch_size,2)
#params_values = NN.train(np.transpose(X), y, epoch, learning_rate,batch_size,2)
# Prediction
Y_train_hat,m = NN.feedforward(np.transpose(X),params_values)

# Accuracy achieved on the train set
acc_train = NN.get_accuracy_value(Y_train_hat, y)
print("Train set accuracy: {:.2f}- ".format(acc_train))
```

```python
# Test the Model
test_file_name = 'data.simple.test.1000.csv'
batch_size = 500

test_ip = np.loadtxt(fname=test_file_name,delimiter=',',skiprows=1,max_rows=batch_size)

# seggregate input and output
y_test = np.array(test_ip[:,2],dtype=int)
y_test = np.where(y_test==2,0,y_test)
X_test = np.array(test_ip[:,:2],dtype=float)
#print(str(X_test),str(y_test))
Y_test_hat,m = NN.feedforward(np.transpose(X_test),params_values)
acc_test = NN.get_accuracy_value(Y_test_hat, y_test)
print("Test set accuracy: {:.2f}- ".format(acc_test))
```
